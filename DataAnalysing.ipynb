```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
df = pd.read_csv('/content/sample_data/immigrants.csv', skiprows=range(0, 3))
df
```

```python
df.describe()
```

```python
indian_row = df.loc[df['Country Name'] == 'India'].iloc[: , 4:69]
# indian_mean = float(indian_row.mean(axis=1).iloc[0])
# indian_row.fillna(indian_mean)
indian_row
```

```python
columns = [i for i in range(1960, 2025)]
rows = indian_row.values.tolist()[0]
plt.plot(columns, rows)
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('Indian immigrants')
plt.show()
```

```python
mean = float(indian_row.mean(axis=1).iloc[0])
std_dev = float(indian_row.std(axis=1).iloc[0])

x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
y = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

cdf = stats.norm.cdf(x, mean, std_dev)
pdf = stats.norm.pdf(x, mean, std_dev)

# Plot CDF
# The CDF graph shows the probability that the number of Indian immigrants is less than or equal to a specific value on the x-axis
plt.figure()
plt.plot(x, cdf, label='CDF')
plt.axvline(mean, color='r', linestyle=':', label='Mean')
plt.xlabel('Number of immigrants')
plt.ylabel('Probability')
plt.title('CDF of Indian immigrants')
plt.legend()
plt.grid(True)
plt.show()

# Plot PDF
# The PDF graph shows the relative likelihood of the number of Indian immigrants taking on a specific value
plt.figure()
plt.plot(x, pdf, label='PDF')
plt.axvline(mean, color='r', linestyle=':', label='Mean')
plt.xlabel('Number of immigrants')
plt.ylabel('Probability Density')
plt.title('PDF of Indian immigrants')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['Mean Immigrants'] = df.iloc[:, 4:69].mean(axis=1)

top_10_countries = df.sort_values(by='Mean Immigrants', ascending=False).head(10)
last_10_countries = df.sort_values(by='Mean Immigrants').head(10)

mean_mean = df['Mean Immigrants'].mean()

plt.bar(top_10_countries['Country Name'], top_10_countries['Mean Immigrants'])
plt.grid(True)
plt.xlabel('Country Name')
plt.ylabel('Mean Immigrants')
plt.title('Top 10 Countries by Mean Immigrants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.bar(last_10_countries['Country Name'], last_10_countries['Mean Immigrants'])
plt.grid(True)
plt.axhline(mean_mean, color='r', linestyle=':', label='Mean')
plt.xlabel('Country Name')
plt.ylabel('Mean of Immigrants')
plt.title('Last 10 Countries by Mean Immigrants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
